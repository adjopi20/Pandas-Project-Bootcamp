```python
#create a column that has the sum of retail disc and coupon disc
project_trans2 = project_trans.assign(
    total_disc = project_trans['RETAIL_DISC'] + project_trans['COUPON_DISC'],
)

project_trans2
```

```python
#create a column that has the percentage discount

project_trans2 = project_trans2.assign(
    percen_disc = (project_trans2['total_disc'] / project_trans2['SALES_VALUE']).abs()
)

project_trans2
```

```python
#if the percentage discount less than 0, set to 0, if it is more than 1 set to 1

project_trans2 = project_trans2.assign(
    percen_disc = project_trans2['percen_disc'].apply(lambda x: 1 if x > 1 else (0 if x < 0 else x))
)

project_trans2
```

```python
#drop the following columns:

project_trans2 = project_trans2.drop(['RETAIL_DISC','COUPON_DISC','COUPON_MATCH_DISC'], axis=1)
```

```python
#PRODUCT ANALYSIS
sales_vs_product = project_trans2.groupby('PRODUCT_ID').agg({'SALES_VALUE':'sum',
                                                                 'QUANTITY' : 'sum'})
sales_vs_product
```

```python
#top 10 product by sales_value in horizontal bar chart
sales_vs_product.sort_values('SALES_VALUE', ascending=False).head(10).plot.barh(y='SALES_VALUE')
```

```python
#top 10 selling item
sales_vs_product = sales_vs_product.sort_values('SALES_VALUE', ascending=False).head(10)
sales_vs_product
```

```python
#avg disc
(project_trans2.query("PRODUCT_ID in @sales_vs_product.index").loc[:, 'total_disc'].sum() /
project_trans2.query("PRODUCT_ID in @sales_vs_product.index").loc[:, 'SALES_VALUE'].sum() 
)
```

```python
#comparing top 10 selling items' discount with avg discount 
sales_vs_product.assign(
    percen_disc = project_trans2['percen_disc'],
    avg_disc =  project_trans2['percen_disc'].mean()
)
```
